# K-means

In this notebook we will explore one of the most famouse clustering algorithms, K-means.

## Data

Instead of importing real data, we are going to create some fake data for educational purposes - the clusters will be quite obvious so we can check visually how well our algorithms are performing.

```python
import numpy as np
from numpy import random, array

# Function that creates k fake clusters for N points
def createClusteredData(N, k):
    random.seed(10)
    pointsPerCluster = float(N)/k
    result = []
    index = []
    for i in range (k):
        xCentroid = random.uniform(20000.0, 200000.0)
        yCentroid = random.uniform(20.0, 70.0)
        for j in range(int(pointsPerCluster)):
            result.append([random.normal(xCentroid, 10000.0), random.normal(yCentroid, 2.0)])
            index.append(i)
    result = array(result)
    index = array(index)
    return result, index
```

For this example, we are going to create 5 clusters with 200 points of data - note that in a real life problem we wouldn't know the number of clusters that we have.

```python
%matplotlib inline
import matplotlib.pyplot as plt
from numpy import random, float
data = createClusteredData(200, 5)
plt.figure(figsize=(8, 6))
plt.scatter(data[0][:,0], data[0][:,1])
plt.show()
```

### K-means

K-means is a very easy to implement clustering algorithm that easily adapts to new examples and generalizes to clusters of different shapes and sizes. However, it is sensitive to outliers and to high dimensionality.

```python
%matplotlib inline

import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import scale

# When using k-means we need to manually choose the number of clusters
model = KMeans(n_clusters=5)

# We scale the data for better results
model = model.fit(scale(data[0]))

plt.figure(figsize=(8, 6))
plt.scatter(data[0][:,0], data[0][:,1], c=model.labels_.astype(float))
plt.show()
```

In practice, we wouldn't know the number of clusters, so let's see what happens for other values of k.

```python
for i in range(2,10):
    model = KMeans(n_clusters=i)
    model = model.fit(scale(data[0]))
    print(model.labels_)
    plt.figure(figsize=(8, 6))
    plt.scatter(data[0][:,0], data[0][:,1], c=model.labels_.astype(float))
    plt.show()
```
