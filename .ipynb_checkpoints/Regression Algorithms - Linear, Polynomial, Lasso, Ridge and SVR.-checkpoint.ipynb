# Regression Algorithms

In this notebook, I am studying the Boston Housing Dataset and using some regression models to predict values for the median value of owner-occupied homes.

### Importing the required libraries

```python
import numpy as np
import matplotlib.pyplot as plt 
import pandas as pd  
import seaborn as sns 
%matplotlib inline
```

### Importing and loading our dataset

```python
from sklearn.datasets import load_boston
boston_dataset = load_boston()
```

### Data exploration

What does the dataset contain?

```python
print(boston_dataset.keys())
```

The Boston Housing dataset contains the data itself, the target, the names of the features and their description as wel as the name of the file. If we wish to access to each of them, we can do so. For example, if we want to know the name of the features:

```python
boston_dataset.feature_names
```

What does each feature mean?

```python
boston_dataset.DESCR
```

'crim': per capita crime rate by town.
'zn': proportion of residential land zoned for lots over 25,000 sq.ft.<br/>
'indus': proportion of non-retail business acres per town.<br/>
'chas':Charles River dummy variable (= 1 if tract bounds river; 0 otherwise).<br/>
'nox': nitrogen oxides concentration (parts per 10 million).<br/>
'rm': average number of rooms per dwelling.<br/>
'age': proportion of owner-occupied units built prior to 1940.<br/>
'dis': weighted mean of distances to five Boston employment centres.<br/>
'rad': index of accessibility to radial highways.<br/>
'tax': full-value property-tax rate per 10,000 dollars.<br/>
'ptratio': pupil-teacher ratio by town. <br/>
'black': 1000(Bk - 0.63)^2 where Bk is the proportion of blacks by town.<br/>
'lstat': lower status of the population (percent).<br/>
'medv': median value of owner-occupied homes in 1000's dollars.<br/>

We now construct our dataframe using Pandas.

```python
df = pd.DataFrame(boston_dataset.data, columns=boston_dataset.feature_names)
df.head()
```

As the target (or label) is missing, we add it manually.

```python
df['MEDV'] = boston_dataset.target
df.head()
```

### Data Cleaning

Now we are going to see if we need to clean the data. First, we check if there are any missing values.

```python
df.info()
```

We are lucky as there is no data missing in our dataframe. Another way to check if there is missing data in our dataframe is using df.isnull().sum(). Now we are going to check if the statistical values make sense.

```python
df.describe()
```

Everything seems reasonable and realistic. For example, for the age column it make sense that the minimun age is almost 3 years and the maximun age is 100. There are no extreme values or negative values.

### Distribution of the label

We are now going to plot the distribution of our label. We use the distplot function from the seaborn library.

```python
fig, axes = plt.subplots(figsize=(18, 6))
sns.distplot(df['MEDV'], label='MEDV').set_title('MEDV distribution', pad=15, size =18)
plt.show()
```

This is a normal distribution with a few outliers. Now we are going to study the correlation between all the features and the label. The correlation coefficient are between -1 and 1. If the value is close to 1, it means that there exist a strong direct correlation between the two variables and if it is close to -1, it means that there exist a strong indirect correlation between the two variables.

```python
fig = plt.figure(figsize=(8, 8), dpi=100)

sns.heatmap(df.corr(), annot=True, square=True, cmap='BuPu', annot_kws={'size': 10}, fmt='.2f', linewidths=.5)   

plt.tick_params(axis='x', labelsize=10, rotation=20)
plt.tick_params(axis='y', labelsize=10, rotation=20)
plt.title('Correlation Matrix', size=18, pad=18)

plt.show()
```

From the correlation matrix we can see that the features that are the most correlated to our label are RM with a correlation on 0.7 and LSTAT with -0.74. We will be interested in these features when fitting a regression model.

A good practice in data exploration is to check for multi-co-linearity. Note that for example the features RAD and TAX have a correlation of 0.91 and therefore we should not select both these features together for training the our regression model. 

Let us study a bit more in depth the relationship bewteen RM, LSTAT and MEDV using a pair-plot.

```python
g = sns.pairplot(df, vars = ['LSTAT', 'RM', 'MEDV'])
g.fig.suptitle('Pair Plots', fontsize=35, y=1.08)

plt.show()
```

The prices of the houses increase linearly as RM increases with a few outliers and as LSTAT decreases not linearly.
The distribution for RM seems to be normal whereas LSTAT doesn't look like it is following a normal distribution.

### Data preparation for training and testing

X are going to be our fetures of interest and Y is going to be our label. the command np.c_ from the numpy library allow us to concatenate the columns LSTAT and RM.

```python
X = pd.DataFrame(np.c_[df['LSTAT'], df['RM']], columns = ['LSTAT','RM'])
Y = df['MEDV']
```

Next, we split the data into our training and testing sets. The models will be trained with 70% of the samples and tested with the remaining 30%. We use train_test_split function provided by the scikit-learn library to split the data.

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.3, random_state=23)
```

We check the size of our training and testing sets.

```python
print(X_train.shape)
print(X_test.shape)
print(Y_train.shape)
print(Y_test.shape)
```

## Linear Regression

We first train our training data using scikit-learn’s LinearRegression.

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

lin_model = LinearRegression()
lin_model.fit(X_train, Y_train)
```

We will evaluate the performace of our model using the RMSE and R2-score.

```python
from sklearn.metrics import r2_score
y_test_predict = lin_model.predict(X_test)
rmse = (np.sqrt(mean_squared_error(Y_test, y_test_predict)))
r2 = r2_score(Y_test, y_test_predict)

print("The model performance for testing set")
print("--------------------------------------")
print('RMSE is {}'.format(rmse))
print('R2 score is {}'.format(r2))
```

Here we plot the actual values of the label in the test data versus the predicted values.

```python
plt.scatter(Y_test, y_test_predict)
plt.show()
```

If the values were predicted perfectly, this would be a straigth line. We can make a prediction of the price of a particular value of RM and LSTAT.

```python
lin_model.predict([[25, 6]])
```

## Polynomial Regression

We first train our training data using Polynomial regression. To do this, we import PolynomialFeatures from sklearn.preprocessing.

```python
from sklearn.preprocessing import PolynomialFeatures
```

We first use a polynomial of degree 2 and follow the steps for the previous algorithm.

```python
polynomial_features= PolynomialFeatures(degree=2)
X_poly = polynomial_features.fit_transform(X)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X_poly, Y, test_size = 0.3, random_state=23)
print(X_train.shape)
print(X_test.shape)
print(Y_train.shape)
print(Y_test.shape)
```

```python
lin_model = LinearRegression()
lin_model.fit(X_train, Y_train)
```

```python
y_test_predict = lin_model.predict(X_test)
rmse = (np.sqrt(mean_squared_error(Y_test, y_test_predict)))
r2 = r2_score(Y_test, y_test_predict)

print("The model performance for testing set")
print("--------------------------------------")
print('RMSE is {}'.format(rmse))
print('R2 score is {}'.format(r2))
```

```python
plt.scatter(Y_test, y_test_predict)
plt.show()
```

```python
lin_model.predict(polynomial_features.fit_transform([[25, 6]]))
```

Let's see what happens if we use polinomials of higher order.

```python
for i in range(2,10):
    polynomial_features= PolynomialFeatures(degree=i)
    X_poly = polynomial_features.fit_transform(X)
    X_train, X_test, Y_train, Y_test = train_test_split(X_poly, Y, test_size = 0.3, random_state=23)
    lin_model = LinearRegression()
    lin_model.fit(X_train, Y_train)
    y_test_predict = lin_model.predict(X_test)
    rmse = (np.sqrt(mean_squared_error(Y_test, y_test_predict)))
    r2 = r2_score(Y_test, y_test_predict)
    
    print(i)
    print("The model performance for testing set")
    print("--------------------------------------")
    print('RMSE is {}'.format(rmse))
    print('R2 score is {}'.format(r2))

    plt.scatter(Y_test, y_test_predict)
    plt.show()
```

We will explore now Lasso and Ridge regression.

Now we use Support Vector Machine algorithm. It's default kernel is rbf.

## Support Vector Machine (Regression)

```python
from sklearn.svm import SVR
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
regr = make_pipeline(StandardScaler(), SVR(C=1, epsilon=1))
regr.fit(X_train, Y_train)
```

```python
y_test_predict = regr.predict(X_test)
rmse = (np.sqrt(mean_squared_error(Y_test, y_test_predict)))
r2 = r2_score(Y_test, y_test_predict)

print("The model performance for testing set")
print("--------------------------------------")
print('RMSE is {}'.format(rmse))
print('R2 score is {}'.format(r2))
plt.scatter(Y_test, y_test_predict)
plt.show()
```

We try now a linear kernel.

```python
regr = make_pipeline(StandardScaler(), SVR(C=0.6, epsilon=1, kernel = 'linear'))
regr.fit(X_train, Y_train)
```

```python
y_test_predict = regr.predict(X_test)
rmse = (np.sqrt(mean_squared_error(Y_test, y_test_predict)))
r2 = r2_score(Y_test, y_test_predict)

print("The model performance for testing set")
print("--------------------------------------")
print('RMSE is {}'.format(rmse))
print('R2 score is {}'.format(r2))

plt.scatter(Y_test, y_test_predict)
plt.show()
```

We try now a polynomial kernel.

```python
regr = make_pipeline(StandardScaler(), SVR(C=2, epsilon=2, kernel = 'poly'))
regr.fit(X_train, Y_train)
```

```python
y_test_predict = regr.predict(X_test)
rmse = (np.sqrt(mean_squared_error(Y_test, y_test_predict)))
r2 = r2_score(Y_test, y_test_predict)

print("The model performance for testing set")
print("--------------------------------------")
print('RMSE is {}'.format(rmse))
print('R2 score is {}'.format(r2))
```

We try now a sigmoid kernel.

```python
regr = make_pipeline(StandardScaler(), SVR(C=0.3, epsilon=0.3, kernel = 'sigmoid'))
regr.fit(X_train, Y_train)
```

```python
y_test_predict = regr.predict(X_test)
rmse = (np.sqrt(mean_squared_error(Y_test, y_test_predict)))
r2 = r2_score(Y_test, y_test_predict)

print("The model performance for testing set")
print("--------------------------------------")
print('RMSE is {}'.format(rmse))
print('R2 score is {}'.format(r2))
```

I have manipulated the hyperparameters in other to get the best R2 score I could get.

The algorithm that gives us the highest score with our choice of hyperparameters is the **Ridge Regression** algorithm.
